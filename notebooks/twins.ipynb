```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
twin_pair_dfs=[]

for file in os.listdir('../data/'):
    if 'twin_pairs' in file:
        df=pd.read_csv(f'../data/{file}')
        unwanted_columns= [col for col in df.columns if 'Unnamed' in col]
        df.drop(columns=unwanted_columns,inplace=True)
        twin_pair_dfs.append(df)
```

```python
df= pd.concat(twin_pair_dfs,axis=1)
```

```python
df.columns[df.isna().mean()==0]
```

```python
y=df['dbirwt_1']-df['dbirwt_0']
x=df['mager8']

plt.subplots(1,2,figsize=(10,8))
plt.subplot(1,2,1)
plt.hist(y)
plt.title('distribution of weights diffence in twins')

plt.subplot(1,2,2)
plt.hist(x)
plt.title('mothers age in octiles')

plt.show()
```
